```python
#function to find the second largest
def secondlargest(li):
    unique=[]
    for n in li:
        if n not in unique:
            unique.append(n)
    unique=sorted(unique,reverse=True)
    if len(unique)>1:
        return unique[1]
    else:
        return -1
secondlargest([2,2,1,1,1])
#secondlargest([2])
```

```python
#fifth least
def fifthleast(li):
    a=[]
    for i in li:
        if i not in a:
            a.append(i)
    a.sort(reverse=True)
    print(a)
    if len(a)>=1:
        return a[4]
    else:
        return -1
fifthleast([1,2,3,4,5])
            
```

```python
def kSmallest(li,k):
    unique=[]
    for i in li:
        if i not in unique:
            unique.append(i)
    unique=sorted(unique)
    if len(unique)>=k:
        return unique[k-1]
    return -1
#kSmallest([3,3,9],2)
kSmallest([],2)
    
```

```python
#using dictionaries
def highestfrequencyElement(li):
    unique={}
    for n in li:
        if n in unique:
            unique[n]+=1
        else:
            unique[n]=1
    print(unique.values())
    #to get the max value
    maxfreq=max(unique.values())
    print(maxfreq)
    maxfreqkeys=[]
    #identify the keys with maximum 
    for item in unique.items():
        
        #item--(key,value)
        if item[1]==maxfreq:
            maxfreqkeys.append(item[0])
    print(maxfreqkeys)
    
    return min(maxfreqkeys)#max(maxfreqkeys) for maximum highest frequency
highestfrequencyElement([1,2,3,2,1,4,4,9])
```
