## Linear Regression

## y=mx+c
* m--slope
* c--constant

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv("headbrain.csv")
a.tail()
```

```python
#collecting x and y
x=a['Head Size(cm^3)'].values
y=a['Brain Weight(grams)'].values
y
```

```python
#calculating the mean of x and y
x_mean=x.mean()
y_mean=y.mean()
print(x_mean,y_mean)
```

```python
#calculating the slope m
a['x-x_mean']=a['Head Size(cm^3)']-x_mean
a['y-y_mean']=a['Brain Weight(grams)']-y_mean
a.head()
```

```python
a['(x-x_mean)*(y-y_mean)']=a['x-x_mean']*a['y-y_mean']
a.head()
```

```python
a["slope"]=a['(x-x_mean)*(y-y_mean)'].sum()/((a['x-x_mean']**2).sum())
a.head()
```

```python
a["constant"]=y_mean-a['slope']*x_mean
a.head()
```

```python
#plotting the values and regression line

#Plotting values and Regression line
#max_x=np.max(x)+2300
#min_x=np.min(x)-2300
#calculating line values x and y
X=np.linspace(min(x)-1000,max(x)+1000,1000)
Y=a['constant'][1]+a['slope'][1]*X   #calculating the regression (y=mx+c)
#ploting line
plt.plot(X,Y,color='yellow',label='Regression Line')
#ploting Scatter Points
plt.scatter(x,y,color='red',label='Scatter Plot')
plt.xlabel('headsize in cm')
plt.ylabel('Brain Weight in grames')
plt.legend()
plt.show()
```
