```python
#function to generate marks data for n students
from random import randint
def generatemarks(n,lb,ub):
    with open('Datafiles/marks.txt','w') as f:
        for i in range(0,int(n)):
            r=randint(int(lb),int(ub))
            f.write(str(r)+'\n')
            print(r,end=" ")
def classaverage():
    sum=0
    count=0
    with open('Datafiles/marks.txt','r') as f:
        for line in f:
            sum=sum+int(line)
            count+=1
        print("class average is",sum//count)
                
def passpercent(lb):
    c=0
    count=0
    
    
    with open('Datafiles/marks.txt','r') as f:
        for line in f:
            count=count+1
            if int(line)>=lb:
                    c+=1
        print(c)        
        pas=(c/count)*100
    print("paspercent:",pas,'%')
def failpercent(lb):
    c=0
    count=0
    with open('Datafiles/marks.txt','r') as f:
        for line in f:
            count=count+1
            if int(line)<lb:
                    c+=1
        print(c)        
        fail=(c/count)*100
    print("failpercent:",fail,'%')
def distniction(lb):
    c=0
    count=0
    with open('Datafiles/marks.txt','r') as f:
        for line in f:
            count=count+1
            if int(line)>=lb:
                    c+=1
        print(c)        
        dist=(c/count)*100
    print("% of distniction:",dist)

def highestmarkfreq(): 
    with open('Datafiles/marks.txt','r') as f:    
        h=[]
        for line in f:
   
                h.append(int(line))
        print("highestmarkfrequency-",max(h),"-->",h.count(max(h)))
def lowestmarkfreq(): 
    with open('Datafiles/marks.txt','r') as f:    
        h=[]
        for line in f:
            h.append(int(line))
        print("lowestmarkfrequency-",min(h),"-->",h.count(min(h)))



while 1:
    print("1.generatemarks\n2.classaverage\n3.passpercent\n4.failpercent\n5.distniction6. highestmarkfreq\n7.lowestmarkfreq")

    n=int(input("enter your choice"))
    
    if n==1:
        
        n,lb,ub=input("enter the boundaries").split(',')
        generatemarks(n,lb,ub)
    elif n==2:
        classaverage()
    elif n==3:
        lb=int(input("enter the boundary value"))
        passpercent(lb)
    elif n==4:
        lb=int(input("enter the boundary value"))
        failpercent(lb)
    elif n==5:
        lb=int(input("enter the boundary value"))
        distniction(lb)
    elif n==6:
        highestmarkfreq()
    elif n==7:
        lowestmarkfreq()
    print("\nenter 1 to continue\n0 to exit")
    p=int(input())
    if p==0:
        break
    elif p==1:
        continue
        
        
        
        
        
        
    
            
        
        
        
```

```python
filename='Datafiles/co.txt'
import re
def phnovalidator(no):
    pattern='[9876][0-9]{9}$|^[0][6-9][0-9]{9}$|^[+][9][1][6-9][0-9]{9}$'
    if re.match(pattern,str(no)):
        return True
        #print("valid number")
    else:
        return False
        #print("invalid number")
    return
def emailval(email):
    pattern='^[0-9a-z][0-9a-z_.]{4,13}[0-9a-z][@][0-9a-z]{3,18}[.][a-z]{2,4}$'
    if re.match(pattern,email):
        #print("valid")
        return True
    else:
        #print("invalid")
        return False
def csvtolist(filename):
    li=[]
    with open(filename,'r') as f:
        for line in f:
            li.append(line.split(','))
    return li


def listofile(li):
    s=''
    for i in li:
        s+=','.join(i)
    return(s)
    

def addcontact(name,no,email):
    with open(filename,'a') as f:
        if not searchcontact(name):
            if phnovalidator(int(no)):
                
                if emailval(email):
                    f.write(name+','+str(no)+','+email+'\n')
                    print("contact added successfully")
        else:
            print("invalid,")
#addcontact("prabhukumari",9963674059,"sandhya4059@gmail.com")
def searchcontact(name):
    fh=csvtolist(filename)
    for line in fh:
        
        if line[0]==name:
            
            #print(line[0],":",line[1],",",line[2])
            print("contact exists")
            return True
    print("contact does not exists" )    
    return False
#searchcontact("kumari")

def searchwithreturn(name):
    fh=csvtolist(filename)
    for line in range(len(fh)):
        
        if fh[line][0]==name:
            
            #print(line[0],":",line[1],",",line[2])
            #print("contact exists")
            return line
        
    return -1
    
        
    

def delcon(name):
    a=searchwithreturn(name)
    print(a)
    if a!=-1:
        fh=csvtolist(filename)
        fh.pop(a)
        fh=listofile(fh)
        with open(filename,'w') as f:
            f.write(fh)
        print("deleted contact  successfully")
    else:
        print("contact not found")
#delcon("prabhukumari")
def listallcontacts():
    with open(filename,'r') as f:
        f=csvtolist(filename)
        for line in f:
            print(line[0])
            print("phoneno:",line[1],end=" ")
            print()
            print("email-id:",line[2],end=" ")
#listallcontacts()
def modifycontact(name,no,email):
    with open(filename,'r') as f:
        i=searchwithreturn(name)
        if i!=-1:
            fh=csvtolist(filename)
            
            for i in range(len(fh)):
                if fh[i][0]==name:
                    fh[i][1]=str(no)
                    fh[i][2]=email
                fa=listofile(fh)
                with open(filename,'w') as f:
                    f.write(fa)
            print("modified successfully")
        else:
            print("contact does not exists to modify")
                    
#modifycontact("prabhu",8121618555,"prabhupinisetti@gmail")

while 1:
    print("1.addcontact\n2.searchcontact\n3.deletecontact\n4.listallcontacts\n5.modifycontact")

    n=int(input("enter your choice"))
    
    if n==1:
        
        name,no,email=input("enter the name,no,email").split(',')
        addcontact(name,no,email)
    elif n==2:
        name=input("enter the name")
        searchcontact(name)
    elif n==3:
        name=input("enter the name")
        delcon(name)
    elif n==4:
        listallcontacts()
    elif n==5:
        name,no,email=input("enter the name,no,email").split(',')
        modifycontact(name,no,email)
    
    print("\n press 1 to continue\n0 to exit")
    p=int(input())
    if p==0:
        break
    elif p==1:
        continue
                
            
    
        
        


        
    
    

```
