#### titanic dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
data=pd.read_csv("titanic.csv")
```

```python
data.head(2)
```

```python
print("no of passengers:"+str(len(data["PassengerId"])))
```

```python
sns.countplot(x='Survived',hue="Sex",data=data)
```

```python
data.isnull().sum()
```

```python
sns.boxplot(x="Pclass",y="Age",data=data)
```

```python
data.drop("Cabin",axis=1,inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
sex=pd.get_dummies(data['Sex'])
sex.head(2)
```

```python
sex=pd.get_dummies(data['Sex'],drop_first=True)
sex.head(2)
```

```python
embark=pd.get_dummies(data['Embarked'])
embark.head(2)
```

```python
embark=pd.get_dummies(data['Embarked'],drop_first=True)
embark.head(2)
```

```python
pcl=pd.get_dummies(data['Pclass'],drop_first=True)
pcl.head(2)
```

```python
data=pd.concat([data,sex,embark,pcl],axis=1)
data
```

```python
data.drop(['Sex','Embarked','PassengerId','Name','Ticket','Pclass'],axis=1,inplace=True)
```

```python
data.head(2)
```

```python
data['Age'].replace(np.NaN,0,inplace=True)
data.head(3)
```

```python
x=data.drop("Survived",axis=1) #INDEPENDENT VARIABLE
y=data["Survived"]
```

```python
from sklearn.model_selection import train_test_split #give model_selection
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
predictions=logmodel.predict(x_test)
predictions
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions)*100)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```
